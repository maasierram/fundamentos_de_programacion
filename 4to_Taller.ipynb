Se necesita un sistema que despliega una tabla de
multiplicar de un número dado por el usuario. ciclo for

```python
numero = int(input("Digite el número deseado: "))

for a in range(11):
  print(a, "x", numero, "=", numero * a)
else:
  print("fin del ciclo for")
```

Se necesita un sistema que pida un sueldo de un
trabajador, si el sueldo es mayor a 655000, no tedrá
ninguna bonificación, pero si es menor tiene bonificación
del 4%. Al final se desea saber el total del sueldo con
o sin bonificación. Preguntar si desea ejecutar de nuevo
el programa con el ciclo while.

```python
pregunta = "si"
while pregunta == "si":
  sueldo = float(input("Digite el salario mensual: "))

  bonificacion = round(sueldo * 0.04, 2)
  total = sueldo + bonificacion

  if sueldo <= 655000:
    estado = "APROBADA"
    print("{}, la bonificacion es= ${:,.0f} y su sueldo sera= {:,.0f}".format(estado, bonificacion, total))
  else:
    estado = "REPROBADA"
    print("{} la bonificacion y su sueldo es= ${:,.0f}".format(estado, sueldo))
  pregunta = input("¿Desea verificar otra salario para el proceso? (si/no) ").lower()
```

Un maestro necesita un sistema para capturar las
calificaciones de 5 parciales de sus alumnos, después
recapturarlas necesita que se despliegue el promedio,
cuando ya no quiera capturar más alumnos, necesita que
se despliegue el promedio general de todos los alumnos
capturados. Preguntar si desea ejecutar de nuevo el
programa con el ciclo while.

```python
pregunta ="si"

while pregunta == "si":
  alumnos = int(input("Digite la cantidad de alumnos: "))
  cal_1 = float(input("Digite la calificacion_1: "))
  cal_2 = float(input("Digite la calificacion_2: "))
  cal_3 = float(input("Digite la calificacion_3: "))
  cal_4 = float(input("Digite la calificacion_4: "))
  cal_5 = float(input("Digite la calificacion_5: "))



  promedio = round((cal_1 + cal_2 + cal_3 + cal_4 + cal_5) / 5, 2)
  total = round(promedio * alumnos/alumnos, 2)
  
  if alumnos < 2:
    print("El promedio de calificaciones es: {}".format(promedio))
  else:
    print("El promedio general es= {}".format(total))
  pregunta = input("¿Desea volver ingresar mas alumnos? (si/no)").lower()
```

Escribir un programa que pregunte al usuario su edad y
muestre por pantalla todos los años que ha cumplido
(desde 1 hasta su edad). ciclo for

```python
edad = int(input("¿Cual es su edad?: "))
for i in range(edad):
  print( 1 + i, "Años")
else:
  print("Se termino el ciclo for")
```
